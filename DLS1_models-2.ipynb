```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil

dataset_path = "/content/dataset"
source = "/content/drive/MyDrive/BoneFractureYolo8"

if not os.path.exists(dataset_path):
    shutil.copytree(source, dataset_path)

data_yaml = os.path.join(dataset_path, "data.yaml")
with open(data_yaml, 'w') as f:
    f.write("train: /content/dataset/train/images\n")
    f.write("val: /content/dataset/valid/images\n")
    f.write("test: /content/dataset/test/images\n")
    f.write("\n")
    f.write("nc: 7\n")
    f.write("names: ['elbow positive', 'fingers positive', 'forearm fracture', 'humerus fracture', 'humerus', 'shoulder fracture', 'wrist positive']\n")

print("Dataset ready")
```

```python
from ultralytics import YOLO

model_fast = YOLO('yolov8n.pt')
model_fast.train(
    data='/content/dataset/data.yaml',
    epochs=150,
    imgsz=640,
    batch=16,
    device=0,
    name='fracture_fast',
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=15,
    translate=0.1,
    scale=0.5,
    fliplr=0.5,
    lr0=0.001,
    lrf=0.01
)
metrics_fast = model_fast.val()
print(f"Fast model mAP@0.5: {metrics_fast.box.map50:.4f}")

import shutil
shutil.copy('/content/runs/detect/fracture_fast/weights/best.pt', '/content/best_fast.pt')
print("Saved to /content/best_fast.pt")
```

```python
from ultralytics import YOLO
model_accurate = YOLO('yolov8m.pt')
model_accurate.train(
    data='/content/dataset/data.yaml',
    epochs=60,
    imgsz=640,
    batch=8,
    device=0,
    name='fracture_accurate',
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=15,
    translate=0.1,
    scale=0.5,
    fliplr=0.5,
    lr0=0.001,
    lrf=0.01
)
```

```python
metrics_accurate = model_accurate.val()
print(f"Accurate model mAP@0.5: {metrics_accurate.box.map50:.4f}")

import shutil
for dirpath, dirnames, filenames in os.walk('/content/runs'):
    for f in filenames:
        if 'accurate' in dirpath and f == 'best.pt':
            shutil.copy(os.path.join(dirpath, f), '/content/best_accurate.pt')
            print("Saved to /content/best_accurate.pt")
```
